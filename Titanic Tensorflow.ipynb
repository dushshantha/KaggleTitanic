```python
#convert male and female to 0 and 1

pd_train["sex_n"] = pd_train["Sex"].apply(lambda x: 1 if x == 'male' else 0)
pd_train
```

```python
# Fill the NaN for Age with Mean

pd_train["Age"] = pd_train["Age"].fillna(pd_train["Age"].mean())

# Fill Nan of Fare
pd_train["Fare"] = pd_train["Fare"].fillna(pd_train["Fare"].mean())
```

```python

pd_train["Embarked"] = pd_train["Embarked"].fillna('S')
pd_train.Embarked.unique()
```

```python
pd_train = pd_train.dropna(how='any', axis=0)
```

```python
# Features 

CATEGORICAL_COLUMNS = ["Pclass", "Sex", "Embarked"]
CONTINUOUS_COLUMNS = ["Age", "Fare", "SibSp"]

def input_fn(df):
  
    #Creates a dictionary mapping from each categorical feature column name (k)
    # to the values of that column stored in a tf.SparseTensor.
    categorical_cols = {k: tf.SparseTensor(
      indices=[[i, 0] for i in range(df[k].size)],
      values=df[k].values,
      dense_shape=[df[k].size, 1])
                      for k in CATEGORICAL_COLUMNS}
    
    continuous_cols = {k: tf.constant(df[k].values)
                     for k in CONTINUOUS_COLUMNS}  
    # Merges the two dictionaries into one.
    feature_cols = dict(continuous_cols.items() | categorical_cols.items())
    # Converts the label column into a constant Tensor.
    label = tf.constant(df['Survived'].values)
    # Returns the feature columns and the label.
    return feature_cols, label

```

```python
#Feature Engineering
#Categorical Columns
gender   = tf.contrib.layers.sparse_column_with_keys(column_name="Sex", keys=["female", "male"])

embarked = tf.contrib.layers.sparse_column_with_keys(column_name="Embarked", keys=["S", "C", "Q"])

#Continuous Columns that I need to make into Bins
#fare = tf.contrib.layers.sparse_column_with_hash_bucket("Fare", hash_bucket_size=100)

# Regular value columns
sibsp = tf.contrib.layers.real_valued_column("SibSp")
age = tf.contrib.layers.real_valued_column("Age")
fare = tf.contrib.layers.real_valued_column("Fare")
pclass_v = tf.contrib.layers.real_valued_column("Fare")

#Bin the age
age_buckets = tf.contrib.layers.bucketized_column(age, boundaries=[8, 25, 30, 35, 40, 45, 50, 55, 60, 65])
pclass   = tf.contrib.layers.bucketized_column(pclass_v, boundaries=[1,2,2])

# Adding crossed Columns 
gender_x_pclass = tf.contrib.layers.crossed_column([gender, pclass], hash_bucket_size=int(1e4))
age_buckets_x_pclass = tf.contrib.layers.crossed_column([age_buckets, pclass], hash_bucket_size=int(1e4))
age_buckets_x_gender = tf.contrib.layers.crossed_column([age_buckets, gender], hash_bucket_size=int(1e4))

```

```python
#Training the model
m = tf.contrib.learn.LinearClassifier(feature_columns=[
  gender, pclass, embarked, fare, sibsp, age_buckets, gender_x_pclass,
  age_buckets_x_pclass, age_buckets_x_gender])
```

```python
# Training Test Split
import numpy as np
msk = np.random.rand(len(pd_train)) < 0.8

train = pd_train[msk]
test  = pd_train[~msk]

print(len(pd_train))
print(len(train))
print(len(test))

def train_input_fn():
  return input_fn(train)

def eval_input_fn():
  return input_fn(test)
```

```python
#model fitting
m.fit(input_fn=train_input_fn, steps=200)
```

```python
results = m.evaluate(input_fn=eval_input_fn, steps=1)
for key in sorted(results):
    print("%s: %s" % (key, results[key]))
```

```python
pred = m.predict(input_fn=eval_input_fn)
for val in pred:
    print(val)
```
